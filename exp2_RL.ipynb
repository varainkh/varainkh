```python
e = np.random.randint(0, 50, 1000)
print(e)

#positive correlation with some noise
f = e + np.random.normal(0, 10, 1000)
print(f)

g = np.random.randint(0, 10, 1000)
print(g)
```

```python
np.corrcoef(e,g)
```

```python
np.corrcoef(f,g)
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.style.use('ggplot')

plt.scatter(e,f,g)
plt.show
```

```python

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(e, f, g)

ax.set_xlabel('E')
ax.set_ylabel('F')
ax.set_zlabel('G')

plt.show()
```

```python
import pandas as pd

df = pd.DataFrame({'a': np.random.randint(0,50,1000)})
df['b'] = df['a'] + np.random.normal(0,10,1000)
df['c'] = 100 - df['a'] + np.random.normal(0,5,1000)
df['d'] = np.random.normal(0,50,1000)

df.corr()
```

```python
pd.plotting.scatter_matrix(df, figsize=(6,6))
plt.show()
```

```python
plt.matshow(df.corr())
plt.xticks(range(len(df.columns)), df.columns)
plt.yticks(range(len(df.columns)), df.columns)
plt.colorbar()
plt.show()
```
